

# 📊 Exam Score Prediction Based on Study Hours



## 1. Business Problem Statement

### 🔹 Background

An academic analytics team wants to understand how **study time impacts student performance**.

Currently, there is no data-driven approach to estimate expected exam scores based on preparation time. This makes it difficult to:

* Guide students effectively
* Set performance benchmarks
* Provide academic intervention support



### 🔹 Objective

Build a **Simple Linear Regression model** that can **predict a student's exam score** based on the **number of hours studied**.



### 🔹 Business Impact

* Data-driven academic performance insights
* Personalized study recommendations
* Early identification of underperformance risk
* Better academic planning and forecasting


## 2. Dataset Description

### 📂 Dataset Source

Student Scores Dataset (Hours vs Scores)



### 📑 Features

| Column Name | Description             | Role            |
| ----------- | ----------------------- | --------------- |
| Hours       | Number of hours studied | Input Feature   |
| Scores      | Exam score obtained     | Target Variable |



### 🎯 Target Variable

**Scores (Continuous Numeric Value)**

This is a regression problem because the output is continuous.


## 3. Problem Type Identification

| Aspect        | Value                    |
| ------------- | ------------------------ |
| Learning Type | Supervised Learning      |
| Algorithm     | Simple Linear Regression |
| Input Feature | Hours                    |
| Output        | Scores                   |
| Relationship  | Linear                   |

> Simple Linear Regression is an appropriate model for this problem.


## ✨ Model Interpretation

### 🔹 Slope (m)

Represents how much the **exam score increases for each additional hour studied**.

### 🔹 Intercept (c)

Represents the **predicted score when study hours = 0**.

This interpretation helps explain results clearly to academic stakeholders.


## 📊 Evaluation Interpretation

### 🔹 MAE (Mean Absolute Error)

Average prediction error in exam score units.

### 🔹 RMSE (Root Mean Squared Error)

Penalizes larger prediction errors more heavily.

### 🔹 R² Score

Indicates how much variation in exam scores is explained by study hours.

Higher R² → Stronger predictive relationship.



````python
# importing the libraries
import numpy as np 
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
````

````python
# step 2 : Reading and loading the datasets
df = pd.read_csv("Study_Hours_vs_Test_Scores.csv")
df.head()
````

````python
# step 03: Exploratory data analysis 
df.info() 
df.isnull().sum()
````

````python
print("No data Cleaning Required")
````

````python
# step 04: Feature Selection 
X = df[['Study_Hours_per_Week']]
y = df['Test_Score_Percentage']
````

````python
# step 05: Analyzing underfitting
plt.scatter(df['Study_Hours_per_Week'],df['Test_Score_Percentage'])
plt.show()
````

````python
# step 06: Split in train n test
from  sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.2, random_state = 40)
````

````python
# step 07: Model training
from sklearn.linear_model import LinearRegression
model = LinearRegression() 
model
````

````python
# step 08: model fitting for training
model.fit(X_train,y_train)
````

````python
# step 09: Predection
y_pred = model.predict(X_test)
````

````python
# step 10: Evaluation 
from sklearn.metrics import r2_score
print('R2 score', r2_score(y_test,y_pred))
````

````python
# step 11: model testing
# take input form user
hours = float(input("Enter Study Hours: "))
hours = np.array([[hours]])
predicted_score = model.predict(hours)
print(f"Predicted Score: {predicted_score[0]:.2f}")
````

````python
# step 12: model saving
import joblib
#save model
joblib.dump(model,"student_score_model.pkl")
print("Model Saved Successfully")
````
