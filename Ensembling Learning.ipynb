```python
x=data.drop('target',axis=1)
y=data['target']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

X_train_scaled=scaler.fit_transform(X_train)
X_test_scaled=scaler.transform(X_test)
```

# Apply the individual model (lg, dt, knn)

# Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
clf1=LogisticRegression()
clf1.fit(X_train,y_train)
clf1.predict(X_test)
from sklearn.metrics import accuracy_score
print('Lg R Acc:', accuracy_score(y_test,clf1.predict(X_test)))
```

# Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
clf2=DecisionTreeClassifier()
clf2.fit(X_train,y_train)
clf2.predict(X_test)
from sklearn.metrics import accuracy_score
print('DT Acc:', accuracy_score(y_test,clf2.predict(X_test)))
```

# KNN

```python
from sklearn.neighbors import KNeighborsClassifier
clf3=KNeighborsClassifier()
clf3.fit(X_train,y_train)
clf3.predict(X_test)
from sklearn.metrics import accuracy_score
print('KN Acc:', accuracy_score(y_test,clf3.predict(X_test)))
```

```python
# META LEARNER ( 3 DEGREE CSE+MECG+CVIL)
from sklearn.ensemble import RandomForestClassifier
meta_clf=RandomForestClassifier(n_estimators=50,random_state=42)
```

# Stacking

note= IN stacking always run whole code at a time without error

Only one meta learner that is Random Forest.

```python
from sklearn.ensemble import RandomForestClassifier
meta_clf=RandomForestClassifier(n_estimators=50,random_state=42)
#--apply stacking--
from sklearn.ensemble import StackingClassifier
Stacking_clf=StackingClassifier(
    estimators=[('lr',clf1),('DT',clf2),('knn',clf3)],
    final_estimator=meta_clf
)
Stacking_clf.fit(X_train,y_train)
y_stack_pred=Stacking_clf.predict(X_test)
print('Stacking Accuracy:',accuracy_score(y_stack_pred,y_test))
```

# Bagging method 01: Adaboost (Adaptive Boosting)

```python
from sklearn.ensemble import AdaBoostClassifier
ada_clf=AdaBoostClassifier()
ada_clf.fit(X_train,y_train)
y_pred_ada=ada_clf.predict(X_test)
from sklearn.metrics import accuracy_score
print('Adaboost accuarcy:',accuracy_score(y_pred_ada,y_test))
```

# Method 02: Gradient Boosting

```python
from sklearn.ensemble import GradientBoostingClassifier
gb_clf=GradientBoostingClassifier()
gb_clf.fit(X_train,y_train)
y_pred_gb=gb_clf.predict(X_test)
from sklearn.metrics import accuracy_score
print('Gradient boosting accuracy:',accuracy_score(y_pred_gb,y_test))
```

# Method 03: XGBoost

```python
import xgboost as xgb
xgb_clf=xgb.XGBClassifier()
xgb_clf.fit(X_train,y_train)
y_pred_xgb=xgb_clf.predict(X_test)
from sklearn.metrics import accuracy_score
print('XGboost accuracy:',accuracy_score(y_pred_xgb,y_test))
```

# Method 04: LightGBM (Light Gradient Boosting Method)

```python
import lightgbm as lgb
lgb_clf=lgb.LGBMClassifier()
lgb_clf.fit(X_train,y_train)
y_pred_lgb=lgb_clf.predict(X_test)
print('XGboost accuracy:',accuracy_score(y_pred_lgb,y_test))
```

# Voting

```python
x=data.drop('target',axis=1)
y=data['target']
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train_scaled=scaler.fit_transform(X_train)
X_test_scaled=scaler.transform(X_test)
```

```python
from sklearn.linear_model import LogisticRegression
clf1=LogisticRegression()
from sklearn.tree import DecisionTreeClassifier
clf2=DecisionTreeClassifier()
from sklearn.neighbors import KNeighborsClassifier
clf3=KNeighborsClassifier()
```

# Hard Voting

```python
from sklearn.ensemble import VotingClassifier
from sklearn.metrics import accuracy_score
voting_hard=VotingClassifier(estimators=[('lr',clf1),('DT',clf2),('knn',clf3)],voting='hard')
voting_hard.fit(X_train,y_train)
y_pred_hard=voting_hard.predict(X_test)
print("Hard Voting Accuracy:",accuracy_score(y_test,y_pred_hard))
```

# Soft Voting

```python
from sklearn.ensemble import VotingClassifier
from sklearn.metrics import accuracy_score
voting_soft=VotingClassifier(estimators=[('lr',clf1),('DT',clf2),('knn',clf3)],voting='soft')
voting_soft.fit(X_train,y_train)
y_pred_soft=voting_soft.predict(X_test)
print("soft Voting Accuracy:",accuracy_score(y_test,y_pred_soft))
```

# Conclusion

Soft Voting is always gives high accuracy compared to hard voting.
