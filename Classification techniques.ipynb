```python
x=data.drop('target',axis=1)
y=data['target']
```

```python
data['target']=(data['target']>0).astype(int)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

X_train_scaled=scaler.fit_transform(X_train)
X_test_scaled=scaler.transform(X_test)
```

# Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
dt_model=DecisionTreeClassifier()
dt_model
```

```python
from sklearn.tree import DecisionTreeClassifier
dt_model=DecisionTreeClassifier(max_depth=5, criterion='gini',random_state=42)
dt_model
# Model fitting:
dt_model.fit(X_train, y_train)
# Model predict:
dt_pred=dt_model.predict(X_test)
# Model evaluation:
from sklearn.metrics import accuracy_score
dt_acc=accuracy_score(y_test,dt_pred)
print(dt_acc)
```

# Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
rf_model=RandomForestClassifier(max_depth=3,random_state=42)
# Model fitting:
rf_model.fit(X_train, y_train)
# Model predict:
rf_pred=rf_model.predict(X_test)
# Model evaluation:
from sklearn.metrics import accuracy_score
rf_acc=accuracy_score(y_test,rf_pred)
print(rf_acc)
```

# KNN

```python
from sklearn.neighbors import KNeighborsClassifier
knn_model=KNeighborsClassifier(n_neighbors=7,metric='euclidean',weights='uniform')
knn_model
# model fitting
knn_model.fit(X_train_scaled, y_train)
#model predict
knn_pred=knn_model.predict(X_test_scaled)
#model evaluation
from sklearn.metrics import accuracy_score
knn_acc=accuracy_score(y_test,knn_pred)
print(knn_acc)
```

# Logistic Regression

Note: LGR always requires scaled data.

```python
from sklearn.linear_model import LogisticRegression
lr_model=LogisticRegression(max_iter=1000, random_state=45)
lr_model
```

```python
# model fitting
lr_model.fit(X_train_scaled, y_train)
#model predict
lr_pred=lr_model.predict(X_test_scaled)
#model evaluation
from sklearn.metrics import accuracy_score
lr_acc=accuracy_score(y_test,lr_pred)
print(lr_acc)
```

note: max_iter and random state are used when accurac is less.

# Naive Bayes

```python
from sklearn.naive_bayes import GaussianNB
nb_model=GaussianNB()
nb_model
```

```python
# Model fitting:
nb_model.fit(X_train, y_train)
# Model predict:
nb_pred=nb_model.predict(X_test)
# Model evaluation:
from sklearn.metrics import accuracy_score
nb_acc=accuracy_score(y_test,nb_pred)
print(nb_acc)
```

note: only use scale data when accuracy is less.

# SVM-SVClassifier

```python
from sklearn.svm import SVC
svm_model=SVC(kernel='rbf',C=1, random_state=42)    # always c=0 or 1
svm_model
```

```python
# Model fitting:
svm_model.fit(X_train, y_train)
# Model predict:
svm_pred=nb_model.predict(X_test)
# Model evaluation:
from sklearn.metrics import accuracy_score
svm_acc=accuracy_score(y_test,svm_pred)
print(svm_acc)
svm_model
```

# Model Comparision

```python
model_accuracies={
    "DT":dt_acc,
    "RF":rf_acc,
    "svm":svm_acc,
    "lr":lr_acc,
    "nb":nb_acc,
    "knn":knn_acc,
}
model_accuracies
```

# Best model

```python
# best model
best_model_name=max(model_accuracies,
                   key=model_accuracies.get)
best_accuracy=model_accuracies[best_model_name]
print(f"\n Best Model: {best_model_name} with accuracy of {best_accuracy:.4f}")
```
